```python
#Loading dataset using Pandas

import pandas as pd

df = pd.read_csv("customer_shopping_behavior.csv")
df
```

```python
df.info()
```

```python
#Summary statistics using describe

df.describe(include="all")
```

```python
#Calculating null value or missing data in dataset

df.isnull().sum()
```

```python
#Checking which row has null value

df[df['Review Rating'].isna()].index
```

```python
#Imputing ssisng value with median value of product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
#Checking null value

df.isna().sum()
```

```python
#Renaming columns in dataset

df.columns = df.columns.str.lower()
df
df.columns = df.columns.str.replace(' ', '_')
df
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
df_copy = df.copy()
```

```python
#Create New age group based on age
df['age_group'] = pd.qcut(df['age'],q=4, labels =['Young Adult', 'Adult', 'Middle-Age', 'Senior'])
df[['age','age_group']]
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quaterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months':30
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].replace(frequency_mapping)
```
